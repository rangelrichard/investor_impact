## Background and Prior Work

The San Diego housing market has undergone significant transformations in recent years, with investors playing an increasingly prominent role. Traditional economic indicators such as employment rates, interest rates, and median household income have historically been reliable predictors of housing market movements. However, the growing presence of investors has introduced new dynamics that may affect market behavior in ways not fully captured by conventional metrics.

Recent analysis by Redfin (2024) highlights the substantial impact of investor activity in San Diego's housing market. Their research indicates that in the second quarter of 2024, investors purchased 23.7% of homes sold in San Diego, positioning the city as the second-highest in the U.S. for investor home purchases, following Miami at 28.5%. This level of investor activity represents a significant portion of market transactions and potentially influences local market dynamics. The study provides crucial baseline data about investor purchase volumes but leaves open questions about the relationship between investor activity and price movements at the zip code level.[1]

The impact of investors on local housing markets extends beyond simple purchase statistics. As reported by CBS 8 San Diego (2024), local officials and real estate experts have expressed both concern and optimism about investor activity in the market. While increased investor presence may signal long-term market strength, it also raises concerns about competition with first-time homebuyers and potential market distortions. San Diego County Supervisor Terra Lawson-Remer has highlighted the important distinction between local "mom-and-pop" landlords and large institutional investors, noting that the latter may have different impacts on community stability and housing affordability.[2]

This research aims to build upon these findings by examining the specific relationship between investor activity and housing price movements at the zip code level, providing a more granular understanding of how different types of investor activity correlate with local market dynamics. By analyzing rates of change between investor purchase percentages and home purchase prices by zip code, we can better understand whether areas with higher investment activity experience accelerated price appreciation compared to areas with less investor presence.

## Stanford data for income

```python
import pandas as pd
import numpy as np
stanford = pd.read_csv('https://docs.google.com/spreadsheets/d/e/2PACX-1vSOVx_YpUQjyXuAFJxTh-NJTtH0WGQle_JNzCcrZcrNsDJFYfBSyfD5BH_ye0_rCPUeuxjDLOoLEDhp/pub?gid=360141403&single=true&output=csv')
stanford.shape
stanford.head(5)
```

As you can see, this data table looks messy with lots of unnecessary columns such as the placeDcid which is basically similar to the placeName(zipcode), I will drop them below.

```python
# Drop unnecessary columns (source links and 'placeDcid' and the additional columns of year)
columns_to_drop = ["placeDcid", "Source:Count_HousingUnit", "Source:Count_HousingUnit_VacantHousingUnit", "Source:Median_Income_Person", 'Date:Count_HousingUnit_VacantHousingUnit', 'Date:Median_Income_Person']
stanford = stanford.drop(columns=columns_to_drop, errors="ignore")
# Rename the column
stanford.rename(columns={
    "placeName": "RegionName", 
    "Date:Count_HousingUnit": "Year",
    "Value:Count_HousingUnit": "Total_Housing_Units",
    "Value:Count_HousingUnit_VacantHousingUnit": "Vacant_Housing_Units",
    "Value:Median_Income_Person": "Median_Income"
}, inplace=True)

# Convert Year column to string for merging purposes
stanford["Year"] = stanford["Year"].astype(str)

stanford.head(5)
```

The Stanford housing & income data table looks well organized right now, however, the same RegionName appear multiple times just because this data table used each year in 2011 to 2023 as a row, which made the overall label complicated for us to incoporate other data source later on. In this case, I decided to pivot it by making each year the columns instead of rows. In this process, I am also able to see some zip code has the missing data on housing unit which remain 0.0 and some income value remain Nan, in order to make a valid data table we had to drop those rows.

```python
# Pivot the Stanford dataset to make Year columns instead of rows
stanford = stanford.pivot(index="RegionName", columns="Year", values=["Total_Housing_Units", "Vacant_Housing_Units", "Median_Income"])

# Flatten the multi-index column names
stanford.columns = [f"{var}_{year}" for var, year in stanford.columns]

# Reset index to bring RegionName back as a column
stanford.reset_index(inplace=True)

# We found some zip code has the value 0 or Nan in it which we have to get rid off.
stanford = stanford.dropna()
stanford = stanford.loc[~(stanford.iloc[:, 1:] == 0).any(axis=1)]

stanford
```

## Zillow Home Value Index 

```python
import pandas as pd
file_path = "https://files.zillowstatic.com/research/public_csvs/zhvi/Zip_zhvi_uc_sfrcondo_tier_0.33_0.67_month.csv"
df = pd.read_csv(file_path)
df.head(5)
```

This data table has zip code all over America, so we have to sort it down to Sandiego only. And for the time period of avg. home price, we will focus it from 2011 to 2023 since it is what we aim for and the Stanford data only has this time period recorded and available.

```python
# Filter for San Diego city and in California only
zillow = df[(df["City"] == "San Diego") & (df["StateName"] == "CA")]
# Keep only columns for 2011 to 2023
date_columns = [col for col in zillow.columns if col.startswith(("2011", "2023"))]
columns_to_keep = ["RegionID", "RegionName", "StateName", "City", "CountyName"] + date_columns
zillow = zillow[columns_to_keep]
zillow = zillow.reset_index(drop=True)

zillow.head(5)
```

## Final Merged Data table
Now is finally the time to merge all the data together. We will first merge the Redfin and Zillow data table together. Once it is done, we will add Stanford housing data in. Since not all zip codes exist in every dataset, we used an "inner" merge, meaning only zip codes present in all three datasets were retained for analysis. After all the merging process has done, you will see the column order in region, share of sales, investiors purchases, all home sales, average home price from 2011 to 2023, total housing unit from 2011 to 2023, vacant housing unit from 2011 to 2023, and the median income per person from 2011 to 2023.

```python
# Ensure zip codes are in string format for consistent merging
redfin["Zip Code"] = redfin["Zip Code"].astype(str)
zillow["RegionName"] = zillow["RegionName"].astype(str)

# Merge on zip code (inner join to keep only matching zip code)
merged_df = pd.merge(zillow, redfin, left_on="RegionName", right_on="Zip Code", how="inner")

# Drop duplicate zip column
merged_df = merged_df.drop(columns=["Zip Code"])

# Drop unnecessary columns
merged_df = merged_df.drop(columns=["Redfin Metro", "CountyName"], errors="ignore")

#Change column order
columns_order = ["RegionID", "RegionName", "StateName", "City", "Share of Sales", "Investor Purchases", "All Home Sales"]
remaining_columns = [col for col in merged_df.columns if col not in columns_order]
merged_df = merged_df[columns_order + remaining_columns]

#Time to merge Stanford Data
# Once again ensure zip codes (RegionName) are in string format for consistent merging
merged_df["RegionName"] = merged_df["RegionName"].astype(str)
stanford["RegionName"] = stanford["RegionName"].astype(str)

# Merge the Stanford dataset with the existing Zillow & Redfin dataset
merged_df = pd.merge(merged_df, stanford, on="RegionName", how="inner")

merged_df.head(10)
```

```python
# Define Zip Code Column & San Diego Zip Codes
zip_column = "RegionName"

san_diego_zip_codes = [
    92101, 92102, 92103, 92104, 92105, 92106, 92107, 92108, 92109, 92110, 92111, 92113, 92114, 
    92115, 92116, 92117, 92119, 92120, 92121, 92122, 92123, 92124, 92126, 92127, 92128, 92129, 
    92130, 92131, 92139, 92140, 92145, 92154, 92173
]


# Convert Zip Codes to String & Filter for San Diego Zip Codes
df[zip_column] = df[zip_column].astype(str)
df = df[df[zip_column].isin(map(str, san_diego_zip_codes))]

# Set Zip Codes as Index
df.set_index(zip_column, inplace=True)
# Clean Column Names (Remove Extra Spaces)
df.columns = df.columns.astype(str).str.strip()
# Identify columns that represent date values
date_columns = [col for col in df.columns if col[:4].isdigit()]
#Filter Data for 2000 to 2024
df_filtered = df[[col for col in date_columns if "2011-01" <= col <= "2023-12"]]
# Convert home values to numeric format
df_filtered = df_filtered.apply(pd.to_numeric, errors='coerce')
# Reshape from wide to long format
df_long = df_filtered.reset_index().melt(id_vars=[zip_column], var_name="Month", value_name="Home_Value")
# Convert Month column to datetime format
df_long["Month"] = pd.to_datetime(df_long["Month"])
# Filter data strictly from 2000 to 2024
df_long = df_long[(df_long["Month"] >= "2011-01-01") & (df_long["Month"] <= "2023-12-31")]
# Sort data by Month & Zip Code
df_long = df_long.sort_values(by=["Month", "RegionName"])
# Group by year and zip code to calculate yearly average home values
df_grouped = df_long.copy()
df_grouped["Year"] = df_grouped["Month"].dt.year  # Extract year
df_grouped = df_grouped.groupby(["Year", "RegionName"])["Home_Value"].mean().reset_index()

print("\n--- Average Home Values by Zip Code & From 2011 to 2023 ---")
print(df_grouped.head(10))  
```

Some things of note that we can also observe is that all regions have an increase in value in a continous upward trend, some regions have factors that have significantly driven up prices that seem to be unrelated to both income and investors, at least at first glance. 

```python
## YOUR CODE HERE
## Starting off by formatting data so we can use it slightly more flexibly
# Creating a new variable to hold data, as well as cleaning the investor % of sales to be usable
pricechange = merged_df
pricechange["Investor_Share_Clean"] = merged_df["Share of Sales"].str.replace("%", "", regex=False).astype(float)
# Formatting zhvi values to be more readily comparable to investor share of sales, percent increase / 10 due to the values being hard to compare
zhvi_change = grouped_df.groupby("RegionName")["Home_Value"].agg(lambda x: ((x.max() / x.min()) - 1) * 10).reset_index()
zhvi_change.rename(columns={"Home_Value": "ZHVI_Increase"}, inplace=True)
# Moving the new values to a proxy, as well as cleaning up repeats
increase = zhvi_change.groupby(["RegionName"])['ZHVI_Increase'].mean().reset_index()
# Adding the value back to the main dataframe for this graph so that I can use it without worrying about dates again
pricechange = merged_df.merge(increase, on="RegionName", how="left")
# Adding a new variable to see how income has changed as a trend, not an exact measurement
pricechange["Median_Change"] = ((pricechange["Median_Income_2023"] / pricechange["Median_Income_2011"]) - 1) * 10
# Moving the specific data we want to graph to it's own data frame
value_summary = pricechange[['RegionName', 'Investor_Share_Clean', 'ZHVI_Increase', "Median_Change"]]
# Rotating the data to make it more readable for EDA comparison
summary_melted = value_summary.melt(id_vars="RegionName", var_name="Metric", value_name="Value")
```

## K-means Clustering of ZIP Code Areas

We are applying K-means clustering to our San Diego ZIP code dataset with the aim to better understand housing market trends. This unsupervised machine learning technique accomplishes the following:

- Partitions ZIP code areas into K distinct clusters based on three economic indicators, investor share, home value appreciation, and median income.
- Minimizes the within-cluster variance (the sum of squared distances between each point and its assigned centroid)
- Creates spatially coherent groups that can help identify regional patterns

The resulting clusters can potentially reveal natural geographic divisions across San Diego, La Jolla, and Del Mar that may correlate with demographic, economic, or other neighborhood characteristics.

## Clustering Analysis: Real Estate Market Segments

This analysis applies K-means clustering to identify distinct market segments across San Diego ZIP codes based on three key real estate indicators:

1. **Total Home Price Appreciation** (% change from first to most recent year)
2. **Investor Activity** (percentage of properties purchased by investors)
3. **Median Household Income** (2023 estimates)

By standardizing these variables and grouping ZIP codes into 3 clusters, we can identify areas with similar market characteristics. This clustering helps reveal patterns in how investor activity relates to appreciation rates and income levels across different neighborhoods, potentially highlighting where institutional investors may be concentrating their activity.

The resulting map visualizes these clusters geographically, allowing us to identify spatial patterns and potential "hot spots" of investor activity within the San Diego region.

```python
import geopandas as gpd
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Load San Diego ZIP code data
sd_zips = gpd.read_file("latest_san_diego_zip_codes.geojson")

# Ensure ZIP codes are strings for merging
sd_zips['ZIP'] = sd_zips['ZIP'].astype(str)

# Calculate Total_Appreciation from your long_df data
first_year = long_df['Year'].min()
recent_year = long_df['Year'].max()

# Calculate first and last year home values
first_year_data = long_df[long_df['Year'] == first_year].groupby('RegionName')['Home_Value'].mean().reset_index()
last_year_data = long_df[long_df['Year'] == recent_year].groupby('RegionName')['Home_Value'].mean().reset_index()

# Merge to calculate appreciation
appreciation_data = first_year_data.merge(
    last_year_data,
    on='RegionName',
    how='inner',
    suffixes=('_first', '_last')
)

# Calculate total appreciation rate
appreciation_data['Total_Appreciation'] = (appreciation_data['Home_Value_last'] / 
                                          appreciation_data['Home_Value_first'] - 1) * 100


# Get Investor_Share_Clean from merged_df
investor_data = merged_df[['RegionName', 'Investor_Share_Clean']].drop_duplicates()

# Get Median_Income_2023 from merged_df
income_data = merged_df[['RegionName', 'Median_Income_2023']].drop_duplicates()

# Create the combined dataset for clustering
cluster_data = appreciation_data[['RegionName', 'Total_Appreciation']].merge(
    investor_data, on='RegionName', how='inner'
).merge(
    income_data, on='RegionName', how='inner'
)

# Ensure RegionName is string
cluster_data['RegionName'] = cluster_data['RegionName'].astype(str)

# Select variables for clustering
cluster_vars = ['Investor_Share_Clean', 'Total_Appreciation', 'Median_Income_2023']

# Check for missing values and fill or drop as appropriate
print(f"Data before handling missing values: {len(cluster_data)} rows")
missing_values = cluster_data[cluster_vars].isnull().sum()
print("Missing values per column:")
print(missing_values)

if missing_values.sum() > 0:
    # Option 1: Drop rows with missing values
    cluster_data = cluster_data.dropna(subset=cluster_vars)
    print(f"Data after dropping missing values: {len(cluster_data)} rows")

# Standardize the data
scaler = StandardScaler()
scaled_data = scaler.fit_transform(cluster_data[cluster_vars])

# Apply KMeans clustering
n_clusters = 3  # Adjust as needed
kmeans = KMeans(n_clusters=n_clusters, random_state=42, n_init=10)
cluster_data['Cluster'] = kmeans.fit_predict(scaled_data)

# Merge clusters with geospatial data
cluster_map = sd_zips.merge(
    cluster_data[['RegionName', 'Cluster']],
    left_on='ZIP',
    right_on='RegionName',
    how='inner'
)

# Define categorical colormap with nicer colors
cluster_colors = {0: "#00A0B0", 1: "#E69F00", 2: "#009E73"}
cluster_map['Color'] = cluster_map['Cluster'].map(cluster_colors)

# Plot the map with categorical colors
fig, ax = plt.subplots(figsize=(16, 13))
cluster_map.plot(
    color=cluster_map['Color'],
    ax=ax,
    edgecolor='black',
    linewidth=0.5
)

# Add ZIP code labels
for idx, row in cluster_map.iterrows():
    centroid = row['geometry'].centroid
    ax.text(
        centroid.x, centroid.y,
        row['RegionName'],
        ha='center', fontsize=10, fontweight='bold',
        bbox=dict(facecolor='white', alpha=0.8, edgecolor='none', pad=2)
    )

# Get cluster descriptions
cluster_profiles = cluster_data.groupby('Cluster')[cluster_vars].mean()

# Create descriptive labels for legend
labels = []
for i in range(n_clusters):
    # Get average values for this cluster
    inv_share = cluster_profiles.loc[i, 'Investor_Share_Clean']
    appreciation = cluster_profiles.loc[i, 'Total_Appreciation']
    income = cluster_profiles.loc[i, 'Median_Income_2023']
    
    # Create a descriptive label
    if i == 0:
        labels.append(f"Cluster {i}: Low Inv. ({inv_share:.1f}%), High App. ({appreciation:.1f}%)")
    elif i == 1:
        labels.append(f"Cluster {i}: Med Inv. ({inv_share:.1f}%), Med App. ({appreciation:.1f}%)")
    else:
        labels.append(f"Cluster {i}: High Inv. ({inv_share:.1f}%), Med App. ({appreciation:.1f}%)")

# Customize legend manually with descriptive labels
legend_patches = [mpatches.Patch(color=color, label=labels[i]) for i, color in cluster_colors.items()]
ax.legend(handles=legend_patches, title="Clusters", loc='upper right')

ax.set_title('San Diego ZIP Code Clusters Based on Real Estate Metrics', fontsize=16)
ax.set_axis_off()
plt.tight_layout()
plt.savefig('san_diego_clusters.png', dpi=300)
plt.show()

# Print cluster profiles
print("\nCluster Profiles:")
print(cluster_profiles)

# Add count of ZIP codes in each cluster
cluster_counts = cluster_data['Cluster'].value_counts().sort_index()
print("\nZip Codes per Cluster:")
print(cluster_counts)

# Print which zip codes belong to each cluster
print("\nZip codes by cluster:")
for cluster_num in sorted(cluster_data['Cluster'].unique()):
    zip_codes = cluster_data[cluster_data['Cluster'] == cluster_num]['RegionName'].tolist()
    print(f"Cluster {cluster_num}: {sorted(zip_codes)}")
```

# Discussion and Conclusion

It is widely accepted that the investor market favors stability, balanced growth, and low risk, which attract investor activity. Another factor is the available market scale combined with relatively moderate barriers to investment. Along with this prediction, investor activity is typically coupled with house appreciation. However, in contrast to previously reported data, our results suggest that investor activity has a diminished direct influence on house prices and may instead be a consequence of house appreciation events.

Through bar plots and K-means clustering, we see consistency among areas from low-income to high-income where house price appreciation exists, but with little connection to investor activity. This could potentially explain the resulting connection that is revealed in clustering approaches. Thus, contrary to the popular view, our findings suggest that house appreciation and investor activity do not follow a simple causal relationship. While investor activity may contribute to short-term price trends, the initiation and long-term appreciation seem to be driven by broader economic factors such as urban development and inflation. Our results suggest that broader economic factors, such as urban development and inflation, are likely to play a larger role in driving initial price appreciation. Lower-income areas may be more sensitive to these factors, creating early appreciation that attracts investors afterward. This challenges the idea that investor activity is the main driver of rising prices, highlighting the importance of local demand and economic policy. While investor activity may boost price trend in short-term, it does not appear to be the main driver of long-term price trends. From 2011 to 2024, the highest appreciation occurred in low-investor, low-income areas, suggesting that demand, supply constraints, and broader economic forces have a stronger influence.

Ultimately, our study challenges the assumption that rising investor activity directly drives housing market growth. Instead, investor activity seems to reflect potential market health and broader economic conditions. These findings suggest that policy makers in order to stabilize housing markets should focus on improving local economic conditions and supply rather than restricting investor activity.
